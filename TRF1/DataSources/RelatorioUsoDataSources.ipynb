```python
import pandas as pd
df = pd.read_csv('./dadosUsoDataSources.csv', sep=',')
df.head()
```

```python
menorData = df['AnoMesDia'].min()
maiorData = df['AnoMesDia'].max()
qtdDias = maiorData - menorData
print("Relatorio de utilizacao dos DataSources ambiente Producao PJe")
print(f"Periodo de {menorData} ate {maiorData}, total de {qtdDias} dias")
```

```python
df.groupby('DataSource')['QtdConexoesCriadasDia'].sum()
```

```python
DataSourceSomatorio = df.groupby(['AnoMesDia', pd.Grouper(key='DataSource')]).filter(lambda x: x['QtdConexoesCriadasDia'].sum() > 0)
DataSourceSomatorio.groupby(['AnoMesDia', pd.Grouper(key='DataSource')]).agg({'QtdConexoesCriadasDia': sum})
```
